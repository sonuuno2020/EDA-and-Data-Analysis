## **Plots**

```python
plt.pie(df["Accepted Offer"].value_counts(),explode=[0,0.1],labels=['Accepted','Not accepted'],autopct='%1.1f%%',shadow=True, startangle=90);
```

```python
plt.pie(df["Received Offer"].value_counts(),labels=['Recived Offfer','Not receive offer'],
        autopct='%0.1f%%',shadow=True, startangle=180,radius=2);
plt.pie(df["Accepted Offer"].value_counts(),labels=['Accepted','Not accepted'],
        autopct='%1.1f%%',shadow=True, startangle=90,radius=1);

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(loc=5)
```

```python
names=['Aman','Vineeta','Namita','Ghazal','Anupam','Ashneer','Peyush']

for x in names:
  plt.figure(figsize=(10,5))
  df[f"{x} Investment Amount"].plot(kind='hist',bins=np.arange(1,max(df[f"{x} Investment Amount"])+1))
  plt.title(f"Number of Investment in lakhs INR by {x}")
  plt.xlabel("Amount")
  plt.ylabel("counts")
```

```python
names=['Aman','Vineeta','Namita','Ghazal','Anupam','Ashneer','Peyush']
debt=[]
for x in names:
  
  debt.append(sum(df[f"{x} Debt Amount"]))
debt
```

```python
plt.pie(debt,labels=names,autopct='%1.1f%%',explode=[0,0,0,0.1,0,0,0],shadow=True, startangle=90);
plt.title("Debt on sharks")
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(loc=4)
```

```python
investments=[]

for x in names:
  investments.append(df[f"{x} Investment Amount"].sum())

plt.ylabel('in INR Lakhs')
plt.bar(names,investments)
plt.title('Investment by sharks')
```

```python
industry=df["Industry"].value_counts()
plt.figure(figsize=(18,6))
plt.bar(industry.index,industry.values)
plt.ylabel('Business belongs to',fontweight='bold').set_fontsize(16)
plt.xlabel('Industries',fontweight='bold').set_fontsize(16)
```

```python
plt.pie(df["Pitchers Average Age"].value_counts(),explode=[0,0,0.1],autopct='%1.1f%%',labels=['Middle','Young',"Old"]);
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(loc=4)
plt.title("Particpant's age distrubution");
```

```python
middle=df[df['Pitchers Average Age']=="Middle"]
young=df[df['Pitchers Average Age']=="Young"]
old=df[df['Pitchers Average Age']=="old"]
receive=df[df["Received Offer"]==1]
val=[]
for y in [middle,young,old]:
  for x in y["Received Offer"].value_counts():
    if x:
      val.append(x)
    else:
      val.append(0)

```

```python

status =['Receive','Not Receive','Receive','Not Receive']
colors = ['#ff6666', '#ffcc99', '#99ff99']
colors_status = ['#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6']

plt.pie(df["Pitchers Average Age"].value_counts(),labels=['Middle','Young',"Old"],colors=colors, startangle=90,frame=True,radius=1.5);
plt.pie(val,labels=status,radius=0.75,startangle=90)
plt.legend(loc=4)
centre_circle = plt.Circle((0,0),0.5,color='black', fc='white',linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.tight_layout()

plt.show()
```
